```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame and it is already loaded with data

# Total Sales Amount by State
sales_by_state = df.groupby('State')['Amount'].sum()
sales_by_state.plot.bar()
plt.title('Total Sales Amount by State')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
# Average Profit by Category
avg_profit_by_category = df.groupby('Category')['Profit'].mean()
avg_profit_by_category.plot.bar(color='green')
plt.title('Average Profit by Category')
plt.xlabel('Category')
plt.ylabel('Average Profit')
plt.show()
```

```python
# Number of Orders by Category
orders_by_category = df['Category'].value_counts()
orders_by_category.plot.bar(color='orange')
plt.title('Number of Orders by Category')
plt.xlabel('Category')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Sales Amount by Payment Mode
sales_by_payment_mode = df.groupby('PaymentMode')['Amount'].sum()
sales_by_payment_mode.plot.bar(color='purple')
plt.title('Sales Amount by Payment Mode')
plt.xlabel('Payment Mode')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# Total Profit by City
profit_by_city = df.groupby('City')['Profit'].sum()
profit_by_city.plot.bar(color='red')
plt.title('Total Profit by City')
plt.xlabel('City')
plt.ylabel('Total Profit')
plt.show()
```

```python
# Quantity Sold by Sub-Category
quantity_by_subcategory = df.groupby('Sub-Category')['Quantity'].sum()
quantity_by_subcategory.plot.bar(color='blue')
plt.title('Quantity Sold by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# Average Order Amount by Customer
average_order_by_customer = df.groupby('CustomerName')['Amount'].mean()
average_order_by_customer.plot.bar(color='cyan', figsize=(30, 15))  # Use figsize for a larger plot if needed
plt.title('Average Order Amount by Customer')
plt.xlabel('Customer')
plt.ylabel('Average Order Amount')
plt.show()
```
